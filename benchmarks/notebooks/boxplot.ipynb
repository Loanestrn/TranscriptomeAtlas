```python
import pandas as pd
import matplotlib.pyplot as plt

retine = '/home/sturny/stageLGM/projet1/GEO/1_input/GSE159357_data_retina_GEO_tidy.csv'
rpe='/home/sturny/stageLGM/projet1/GEO/1_input/GSE146887_tidy.csv'


dfr = pd.read_csv(retine)
dfrpe=pd.read_csv(rpe)

column_prefixes= [col.split('_')[0] for col in dfr.columns]

uniques_prefixes= list(dict.fromkeys(column_prefixes))
colors_map=dict(zip(uniques_prefixes, plt.cm.tab10.colors))
box_colors=[colors_map[prefix] for prefix in column_prefixes]
box= dfr.plot.box(vert=False, showfliers=False, patch_artist=True)

for patch, color in zip(box.artists, box_colors):
    patch.set_facecolor(color)
plt.title('Boxplot retine with color')

```

```python
import pandas as pd
import matplotlib.pyplot as plt

rpe = '/home/sturny/stageLGM/projet1/GEO/1_input/GSE146887_tidy_correct.csv'


dfrpe = pd.read_csv(rpe, decimal=',')


column_prefixes = [col.split('_')[0] for col in dfrpe.columns]
uniques_prefixes = list(dict.fromkeys(column_prefixes))
colors_map = dict(zip(uniques_prefixes, plt.cm.tab10.colors))
box_colors = [colors_map[prefix] for prefix in column_prefixes]


box = dfrpe.plot.box(vert=False, showfliers=False, patch_artist=True)


for patch, color in zip(box.artists, box_colors):
    patch.set_facecolor(color)

plt.title('Boxplot RPE with Colors')
plt.tight_layout()
plt.show()
```
